```python
from sklearn.preprocessing import MinMaxScaler
MMS = MinMaxScaler()
new_data[new_data.columns] = MMS.fit_transform(new_data)
```

```python
new_data
```

```python
new_data.shape
```

```python
# Selecting 80 % for training and 20 % for testing

training_size = round(len(new_data) * 0.80)
training_size
```

```python
train_data = new_data[:training_size]
test_data  = new_data[training_size:]

train_data.shape, test_data.shape
```

```python
def create_sequence(dataset):
                sequences = []
                labels = []
                start_idx = 0
                for stop_idx in range(50,len(dataset)):                   # Selecting 50 rows at a time
                    sequences.append(dataset.iloc[start_idx:stop_idx])
                    labels.append(dataset.iloc[stop_idx])
                    start_idx += 1
                return (np.array(sequences),np.array(labels))
```

```python
train_seq, train_label = create_sequence(train_data)
test_seq, test_label = create_sequence(test_data)
```

```python
train_seq.shape, train_label.shape, test_seq.shape, test_label.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM, Bidirectional
```

```python
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape = (train_seq.shape[1], train_seq.shape[2])))

model.add(Dropout(0.1))
model.add(LSTM(units=50))

model.add(Dense(2))

model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mean_absolute_error'])

model.summary()
```

```python
model.fit(train_seq, train_label, epochs=80,validation_data=(test_seq, test_label), verbose=1)
```

```python
# Save the model to a file
model.save('/content/drive/MyDrive/project_jazz/SPM/my_model.h5')
```

```python
test_predicted = model.predict(test_seq)
test_predicted[:5]
```

```python
test_inverse_predicted = MMS.inverse_transform(test_predicted) # Inversing scaling on predicted data
test_inverse_predicted[:5]
```

```python
concat_data = pd.concat([new_data.iloc[-152:].copy(),pd.DataFrame(test_inverse_predicted,columns=['open_predicted','close_predicted'],index=new_data.iloc[-152:].index)], axis=1)
```

```python
concat_data[['open','close']] = MMS.inverse_transform(concat_data[['Open','Close']]) # Inverse scaling
```
